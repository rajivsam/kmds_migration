## Prompt for the data download notebook

1. read the zip file in the data dir    ectory with pandas
2. read the data dictionary in the data_dictionary directory with pandas
3. display the head for the data and the data dictionary    

```python
import pandas as pd
import zipfile
import glob

zip_files = glob.glob('../data/*.zip')
if not zip_files:
    raise FileNotFoundError('No zip file found in data directory')
zip_path = zip_files[0]

with zipfile.ZipFile(zip_path) as z:
    csv_files = [f for f in z.namelist() if not f.endswith('/')]
    if not csv_files:
        raise FileNotFoundError('No file found inside the zip archive')
    df = pd.read_csv(z.open(csv_files[0]))

dict_files = glob.glob('../data_dictionary/*.csv')
if not dict_files:
    raise FileNotFoundError('No CSV file found in data_dictionary directory')
data_dict = pd.read_csv(dict_files[0])

display(df.head())
display(data_dict.head())
```

## Instructions for the data file 
1. Include the following columns in the data file - number, opened_at, assignment_group, resolved_at, closed_at.
2. Make the number and assignment_group columns to be categorical (string) data types. The others are datettime
3. subset the data columns to step 1 and save it as itsm_group_analysis.csv in the data directory under the parent directory of this notebook   .

```python
cols = ['number', 'opened_at', 'assignment_group', 'resolved_at', 'closed_at']
df_subset = df[cols].copy()

df_subset['number'] = df_subset['number'].astype('category')
df_subset['assignment_group'] = df_subset['assignment_group'].astype('category')

for dt_col in ['opened_at', 'resolved_at', 'closed_at']:
    df_subset[dt_col] = pd.to_datetime(
        df_subset[dt_col].astype(str),
        dayfirst=True,
        errors='coerce'
    )

output_path = '../data/itsm_group_analysis.csv'
df_subset.to_csv(output_path, index=False)

display(df_subset.head())
```
